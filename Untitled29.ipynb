```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('train_data.csv')
test = pd.read_csv('test_data.csv')
```

```python
train.head()
```

```python
# For train dataframe
columns_to_drop_train = [col for col in train.columns if col.startswith('Unnamed:0') or col == 'Id']
train = train.drop(columns_to_drop_train, axis=1)

# For test dataframe
columns_to_drop_test = [col for col in test.columns if col.startswith('Unnamed:0') or col == 'Id']
test = test.drop(columns_to_drop_test, axis=1)
```

```python
train.head()
```

```python
train_data = train.values
Y = train_data[:,-1].reshape(train_data.shape[0],1)
X = train_data[:,:-1]
```

```python
test_data = test.values
Y = test_data[:,-1].reshape(test_data.shape[0],1)
X = test_data[:,:-1]
```

```python
X = np.vstack((np.ones(X.shape[0]),X.T)).T
test_X = np.vstack((np.ones(X.shape[0]),X.T)).T
```

```python
def model(X,Y, learning_rate, iterations):
  m = Y.size
  theta = np.zeros((X.shape[1],1))
  cost_list = []
  for i in range(iterations):
    y_pred = np.dot(X,theta)
    cost = (1/(2*m))*np.sum(np.square(y_pred-Y))
    d_theta = (1/m)*np.dot(X.T,(y_pred-Y))
    theta = theta - learning_rate*d_theta
    cost_list.append(cost)
    if (i%(iterations/10) == 0):
      print("Cost is:",cost)
  return theta, cost_list
```

```python
iterations = 30000
learning_rate = 0.000000008
theta, cost_list = model(X,Y,learning_rate = learning_rate,iterations = iterations)
```

```python
plt.plot(cost_list)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

```python
y_pred = np.dot(X,theta)

error = (1/X.shape[0]) * np.sum(np.abs(y_pred - Y))
print (error)
```

```python
print("Test error is:", round(error, 4)*100,"%")
print("Test accuracy is:", (1-error)*100,"%")
```
